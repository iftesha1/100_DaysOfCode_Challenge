```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('concrete_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(['Strength'],axis=1)
y=df['Strength']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in x_train.columns:
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    sns.distplot(x_train[i])
    plt.title('PDF')

    plt.subplot(122)
    stats.probplot(x_train[i], dist="norm", plot=plt)
    plt.title('QQ Plot')

    plt.show()
    
```

```python
bct= PowerTransformer(method='box-cox')
```

```python
x_train_transformed=bct.fit_transform(x_train+0.000001)
x_test_transformed=bct.transform(x_test+0.000001)
```

```python
lamba_values=pd.DataFrame({'cols':x_train.columns,'box_cox_lambdas':bct.lambdas_})
lamba_values
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)

print('The r2 score before transformation is',r2_score(y_test,y_pred))
print('The r2 score before transformation with cross-validation is',np.mean(cross_val_score(lr,x,y,cv=10,scoring='r2')))
```

```python
x_transformed=bct.fit_transform(x+0.000001)
lr.fit(x_train_transformed,y_train)
y_pred2=lr.predict(x_test_transformed)

print('The r2 score After transformation is',r2_score(y_test,y_pred2))
print('The r2 score After transformation with cross-validation is',np.mean(cross_val_score(lr,x_transformed,y,cv=10,scoring='r2')))
```

```python
x_train_transformed=pd.DataFrame(data=x_train_transformed,columns=x_train.columns)
for i in x_train.columns:
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    sns.distplot(x_train[i])
    plt.title('PDF before Transformation')

    plt.subplot(122)
    sns.distplot(x_train_transformed[i])
    plt.title('PDF After Transformation')

    plt.show()
```
