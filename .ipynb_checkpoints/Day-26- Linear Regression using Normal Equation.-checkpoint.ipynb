## Linear Regression using Normal Equation


A linear model makes a prediction by simply computing a weighted sum of the input features, plus a constant called the bias term (also called the intercept term).


Where y is the predicted value, n is the number of features, x is the feature value and θ  is the model parameter.

To find the value of θ that minimizes the cost function (or the mean squared error function), there is a closed form solution in other words, a mathematical equation that gives the result directly. This is called the Normal Equation given below.




```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=2*np.random.rand(100,1)
y=4+3*X+np.random.rand(100,1)
```

```python
print(X)
print(y)
```

```python
plt.scatter(X,y)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new
```

```python
 X_new_b = np.c_[np.ones((2, 1)), X_new] # add x0 = 1 to each instance
```

```python
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "g.")
plt.axis([0, 2, 0, 15])
plt.show()
```

We can see that our prediction line almost touches every points of the training dataset. So we can say model has fited well.

## Fiting the model using Sk-learn

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

These parameter values are almost equal to the value we got using normal equation.

```python
lr.predict(X_new)
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```
