```python
import pandas as pd
import numpy as np
```

```python
date = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day34-handling-date-and-time/orders.csv')
time = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day34-handling-date-and-time/messages.csv')
```

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#Extract Year

date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
#Extract Month

date['date_month'] = date['date'].dt.month

date.head()
```

```python
#Extract Month Name

date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
#Extract Day

date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week

date['week_no'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 'yes','No')

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
