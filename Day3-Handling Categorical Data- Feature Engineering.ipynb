## 1.Ordinal Encoding ( Peformed on Ordinal categorical features)

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('customer.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('purchased',axis=1)
Y=df['purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10,stratify=Y)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
df['purchased'].value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
x_train_transformed=oe.fit_transform(x_train)
x_test_transformed=oe.transform(x_test)
```

```python
x_train.head(3)
```

```python
x_train_transformed[:3,:]
```

## 3. One Hot Encoding (Performed on Nominal Categorical Features)

```python
df2 =pd.read_csv('cars.csv')
```

```python
df2.head()
```

```python
df2.brand.value_counts()
```

```python
print(df2['fuel'].value_counts(),'\n')
print(df2['owner'].value_counts())
```

```python
# 1.Using Pandas
pd.get_dummies(df2,columns=['fuel','owner'])
```

```python
# Creating (n-1) feature coulumns
pd.get_dummies(df2,columns=['fuel','owner'],drop_first=True)
```

```python
# Categories with many unique values
counts=df2['brand'].value_counts()
replace = counts[counts <= 100].index
replace
```

```python
pd.get_dummies(df2['brand'].replace(replace,'Not frequents'))
```

```python
df2.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df2.iloc[:,0:4],df2.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
# Using Sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse=False)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new=ohe.transform(X_test[['fuel','owner']])
```
